# Create a Snarky Headline for a website

**The point:** the structure of the code does not change. Only the prompts change. We reuse `messages_for` and the same OpenAI call, just with a new instruction.

```python
# Step 1: Create your prompts

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and writes a short, snarky, humorous truthful headline, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt_prefix = """
Here are the contents of a website.
Provide a short, snarky, humorous truthful headline of this website.
"""
```

## Call scraper.py with function: fetch_website_contents()

```python
# Re-use 3 functions from day1.ipynb just rename and change a bit.

# Step 2: Make the messages list
# messages = [] # fill this in

def messages_for(website): 
    """Build the [system, user] message list expected by the OpenAI API"""
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]


def summarize(url):
    website = fetch_website_contents(url)
    # Step 3: Call OpenAI:  response = ...
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(website)
    )
    # Return the response
    return response.choices[0].message.content


# A function to display this nicely in the output, using markdown
def display_summary(url):
    summary = summarize(url)
    # Step 4: print the result
    display(Markdown(summary))
```

```python
display_summary(ed)
```
